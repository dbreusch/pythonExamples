Create a file with data from closest grid point
===

Import required modules and set file names
----

```python
from netCDF4 import Dataset,date2num
import pandas as pd
import numpy as np

%load_ext autoreload
%autoreload 2
from model import Model
from GCNet import GCNet
```

Period-specific data
----

```python
branch = "historical"
yr1 = "1996"
yr2 = "2005"
code = "B20TRC5CNBDRD"

# branch = "rcp85"
# yr1 = "2071"
# yr2 = "2080"
# code = "BRCP85C5CNBDRD"

# mon = 7  # NOT USED
nEns = 30

suff = str(yr1)+"0101-"+str(yr2)+"1231"
```

File templates etc
---

```python
# modelDataFN = "daily/b.e11.B20TRC5CNBDRD.f09_g16.001.cam.h1.TREFHT.19960101-20051231_sub.nc"
modelDataTemplate = branch+"/daily/b.e11.%s.f09_g16.%03d.cam.h1.TREFHT.%s0101-%s1231_sub.nc"
modelMetaFN = "cesmle_geog_sub.nc"
newNCFN = "cesmle_tas_closest_"+suff+".nc"
nEns = 30
```

Load/define CESM LE metadata
----

```python
M = Model(modelMetaFN)
M.loadMeta()

minLon = -70 + 360
maxLon = -15 + 360
minLat = 58
maxLat = 86.5
M.setDomain( (minLat, maxLat, minLon, maxLon) )
```

Load AWS metadata
----

```python
A = GCNet("site_info.nc")
A.loadMeta()
nSites = A.getNSites()
```

Find closest grid points
-----

```python
C = []
for awsIX in (range(nSites)):
# for awsIX in ((0,1)):
    awsLatLon = ( A.getLat( awsIX ), A.getLon( awsIX ))
    closest = M.closestPoints( awsLatLon )
    C.append( closest )
```

Process TREFHT data
====

```python
for ens in range(nEns):
# for ens in ([0]):
    enss = ens+1
    modelDataFN = modelDataTemplate % ( code, enss, yr1, yr2 )
    print modelDataFN
    D = Model(None, modelDataFN)

    tas = np.array( D.loadData( "TREFHT" ) )
    nRec, nLat, nLon = tas.shape
    print nRec, nLat, nLon

    if ens < 1:
        time, timeCF = D.loadData( "time" )
        Tnew = np.empty( (nEns, nRec, nSites), dtype = 'float' )
    print time[0], time[-1]
    
    for awsIX in (range(nSites)):
    # for awsIX in ((0,1)):
    # for awsIX in ([0]):
        """ Basic AWS info """
        awsName = A.getName( awsIX )
        awsLatLon = ( A.getLat( awsIX ), A.getLon( awsIX ))

        closest = C[awsIX][0]
        dx = "%.1f km" % closest[-1]
        print awsName, dx
#         print Tnew[0][0][0]
        Tnew[ens,:,awsIX] = tas[:,closest['ixLat'],closest['ixLon'] ]
#         print Tnew[0][0][0]
```

Create a netCDF file for the new data
-----
